```python
import math

# -----------------------------
# Activation function
# -----------------------------
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def sigmoid_derivative(output):
    return output * (1 - output)

# -----------------------------
# Input layer
# -----------------------------
x1 = 0.35
x2 = 0.9
X = [x1, x2]

# -----------------------------
# Target output
# -----------------------------
target = 0.5
learning_rate = 1

# -----------------------------
# Weight initialization
# -----------------------------
# Input → Hidden
w13, w23 = 0.1, 0.6
w14, w24 = 0.4, 0.6

# Hidden → Output
w35, w45 = 0.3, 0.9

# -----------------------------
# Forward propagation
# -----------------------------
net_h1 = x1*w13 + x2*w23
net_h2 = x1*w14 + x2*w24

h1 = sigmoid(net_h1)
h2 = sigmoid(net_h2)

net_o = h1*w35 + h2*w45
output = sigmoid(net_o)

print("Network Output:", round(output, 4))

# -----------------------------
# Loss function (MSE)
# -----------------------------
loss = (target - output) ** 2
print("Loss Error:", round(loss, 4))  # ≈ 0.1822

# -----------------------------
# Backpropagation
# -----------------------------
# Output layer error term
delta_o = (target - output) * sigmoid_derivative(output)

# Hidden layer error terms
delta_h1 = sigmoid_derivative(h1) * delta_o * w35
delta_h2 = sigmoid_derivative(h2) * delta_o * w45

# -----------------------------
# Weight updates
# -----------------------------
# Hidden → Output
w35 += learning_rate * delta_o * h1
w45 += learning_rate * delta_o * h2

# Input → Hidden
w13 += learning_rate * delta_h1 * x1
w23 += learning_rate * delta_h1 * x2
w14 += learning_rate * delta_h2 * x1
w24 += learning_rate * delta_h2 * x2

# -----------------------------
# Updated weights
# -----------------------------
print("\nUpdated Weights after Backpropagation:")
print("w13 =", round(w13, 4))
print("w23 =", round(w23, 4))
print("w14 =", round(w14, 4))
print("w24 =", round(w24, 4))
print("w35 =", round(w35, 4))
print("w45 =", round(w45, 4))
```
