```python
import tensorflow as tf
import numpy as np

# -----------------------------
# 1. Training data
# -----------------------------
X = np.array([[1.0]], dtype=np.float32)
y = np.array([[2.0]], dtype=np.float32)

# -----------------------------
# 2. Build the model
# -----------------------------
model = tf.keras.Sequential([
    tf.keras.layers.Dense(3, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(1)  # linear output for regression
])

# -----------------------------
# 3. Compile the model
# -----------------------------
model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
    loss=tf.keras.losses.Huber(delta=1.0)
)

# -----------------------------
# 4. Train the model
# -----------------------------
history = model.fit(
    X, y,
    epochs=100,
    verbose=1
)

# -----------------------------
# 5. Final prediction
# -----------------------------
prediction = model.predict(X)
print("Final prediction:", prediction[0][0])
```

```python
import tensorflow as tf
import numpy as np

# -----------------------------
# 1. Data
# -----------------------------
X = np.array([[1.0]], dtype=np.float32)
y = np.array([[2.0]], dtype=np.float32)

# -----------------------------
# 2. Define input
# -----------------------------
inputs = tf.keras.Input(shape=(1,))

# -----------------------------
# 3. Define layers
# -----------------------------
hidden = tf.keras.layers.Dense(3, activation='relu')(inputs)
outputs = tf.keras.layers.Dense(1)(hidden)

# -----------------------------
# 4. Create model
# -----------------------------
model = tf.keras.Model(inputs=inputs, outputs=outputs)

# -----------------------------
# 5. Compile
# -----------------------------
model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
    loss=tf.keras.losses.Huber(delta=1.0)
)

# -----------------------------
# 6. Train
# -----------------------------
model.fit(X, y, epochs=100, verbose=1)

# -----------------------------
# 7. Predict
# -----------------------------
print("Prediction:", model.predict(X)[0][0])
```
