```python
import numpy as np

# Input data
X = np.array([[1], [2], [3], [4]])   # shape (4,1)
y = np.array([[2], [4], [6], [8]])   # shape (4,1)

# Initialize weights and bias
np.random.seed(0)
W1 = np.random.randn(1, 1)
b1 = np.zeros((1, 1))

# Hyperparameters
learning_rate = 0.01
epochs = 1000

# Training loop
for epoch in range(epochs):
    # -------- Forward Propagation --------
    Z = np.dot(X, W1) + b1          # Linear
    y_pred = Z                      # No activation (linear)

    # -------- Loss (Mean Squared Error) --------
    loss = np.mean((y - y_pred) ** 2)

    # -------- Backpropagation --------
    dLoss_dy = -2 * (y - y_pred) / len(y)
    dLoss_dW = np.dot(X.T, dLoss_dy)
    dLoss_db = np.sum(dLoss_dy)

    # -------- Update Parameters --------
    W1 -= learning_rate * dLoss_dW
    b1 -= learning_rate * dLoss_db

    if epoch % 200 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

print("\nFinal Weight:", W1)
print("Final Bias:", b1)
```
