Perform element wise addition on two(2,3) tensors filled with random values

```python
import tensorflow as tf

# Create two (2, 3) tensors with random values
A = tf.random.uniform((2, 3))
B = tf.random.uniform((2, 3))

# Element-wise addition
C = tf.add(A, B)        # or simply: C = A + B

print("Tensor A:\n", A)
print("\nTensor B:\n", B)
print("\nElement-wise Addition (A + B):\n", C)
```

```python
import tensorflow as tf

# Step 1: Create two (2,3) tensors with random values
tensor1 = tf.random.uniform(shape=(2, 3))
tensor2 = tf.random.uniform(shape=(2, 3))

# Step 2: Perform element-wise addition
result = tensor1 + tensor2   # or tf.add(tensor1, tensor2)

# Step 3: Display the result
print("Tensor 1:\n", tensor1)
print("\nTensor 2:\n", tensor2)
print("\nElement-wise Addition Result:\n", result)
```

```python
import numpy as np

# Create two (2,3) arrays with random values
A = np.random.rand(2, 3)
B = np.random.rand(2, 3)

# Element-wise addition
C = A + B        # or np.add(A, B)

# Display results
print("Array A:\n", A)
print("\nArray B:\n", B)
print("\nElement-wise Addition (A + B):\n", C)
```
