```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# 1. Load pre-trained YOLOv8 model
model = YOLO("yolov8n.pt")   # downloads automatically first time

# 2. Give your input image (JUST the image with objects)
image_path = "/content/object2.jpeg"   # ← change this to your file
# You can upload the image to Colab and check the name with: !ls

# 3. Run object detection
# conf=0.5 means: only keep detections with probability ≥ 0.5
results = model(image_path, conf=0.5)

# 4. Get image with boxes drawn by YOLO
annotated_img = results[0].plot()          # numpy array with boxes + labels

# 5. Show the result in the output section
annotated_img = cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 6))
plt.imshow(annotated_img)
plt.axis("off")
plt.show()
```
